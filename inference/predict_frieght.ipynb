```python
from pathlib import Path

import joblib
import pandas as pd

MODEL_PATH = "ML_project/frieght_cost_prediction/models/predict_freight_model.pkl"


def load_model(model_path: str = MODEL_PATH):
    """Load the trained freight cost prediction model."""
    candidates = [
        Path(model_path),
        Path("../frieght_cost_prediction/models/predict_freight_model.pkl"),
        Path("../frieght_cost_prediction/models/freight_model.pkl"),
        Path("ML_project/frieght_cost_prediction/models/freight_model.pkl"),
    ]

    model_file = next((p for p in candidates if p.exists()), None)
    if model_file is None:
        raise FileNotFoundError("Could not find a trained freight model in expected locations.")

    return joblib.load(model_file)


def predict_freight_cost(input_data):
    """Predict freight cost for new vendor invoices.

    Parameters
    ----------
    input_data : dict or list[dict]
        Input feature values for the model.

    Returns
    -------
    pandas.DataFrame
        Input data with an added Predicted_Freight column.
    """
    model = load_model()
    input_df = pd.DataFrame(input_data)
    input_df["Predicted_Freight"] = model.predict(input_df).round(2)
    return input_df


if __name__ == "__main__":
    # Example inference run (local testing)
    sample_data = {
        "Quantity": [2500, 1200, 500],
        "Dollars": [18500, 9000, 3000],
    }

    prediction = predict_freight_cost(sample_data)
    print(prediction)
```
